```python

# Automate Market Data Collection


# List of stock tickers we want to collect
tickers = ["AAPL", "MSFT", "NVDA", "AMZN", "TSLA", "GOOGL"]

# Empty list where we will store each cleaned dataset
all_stock_data = []

# Repeat the same process for every ticker
for ticker in tickers:

    # Download historical market data
    stock_data = yf.download(
        ticker,
        start="2025-01-01",
        end="2026-01-01",
        progress=False
    )

    # Remove the second column level created by Yahoo Finance
    stock_data.columns = stock_data.columns.get_level_values(0)

    # Convert Date from index into a normal column
    stock_data = stock_data.reset_index()

    # Remove the remaining column index name
    stock_data.columns.name = None

    # Add business metadata
    stock_data["Ticker"] = ticker
    stock_data["DataSource"] = "Yahoo Finance"

    # Store the cleaned dataset in the list
    all_stock_data.append(stock_data)

print("Data collection completed.")
```

```python
# Combine All Stock Datasets


# Combine all DataFrames stored in the list
stock_market_data = pd.concat(
    all_stock_data,
    ignore_index=True
)

# Display the first 10 rows
stock_market_data.head(10)
```

```python
# ==========================================================
# Step 4: Connect Python to SQL Server
#
# Business Purpose:
# Establish a connection between Python and SQL Server.
# This allows Python to load validated market data directly
# into the database without manual CSV imports.
# ==========================================================

from sqlalchemy import create_engine
from urllib.parse import quote_plus

# SQL Server connection settings
server = "localhost"
database = "MarketDataWarehouse"

# ODBC Driver
driver = quote_plus("ODBC Driver 17 for SQL Server")

# Create SQLAlchemy engine
engine = create_engine(
    f"mssql+pyodbc://@{server}/{database}"
    f"?driver={driver}"
    "&trusted_connection=yes"
)

print("Connection created successfully!")
```

```python

# Step 5: Test the SQL Server Connection


# Simple SQL query
query = "SELECT DB_NAME() AS CurrentDatabase;"

# Execute the query
result = pd.read_sql(query, engine)

# Display the result
result
```

```python

# Load the Instruments Dimension Table

instruments.to_sql(
    name="Instruments",
    con=engine,
    if_exists="append",
    index=False
)

print("Instruments table loaded successfully.")
```
